```python
import requests, bs4
import pandas as pd
headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10.14; rv:71.0) Gecko/20100101 Firefox/71.0'}

targetUrl = "https://www.timeanddate.com/holidays/south-korea/?hol=1"
resp = requests.get(targetUrl)
html = resp.text
bs = bs4.BeautifulSoup(html, "html.parser")

Tables = bs.find(name="table", attrs={ "id" : "holidays-table"})
# tagname이 table이고 tagid가 holidays-table인 부분을 발췌해서 Tables변수에 담았다.

Tbodys = Tables.find(name="tbody")
# Tables에서 tagname이 tbody부분만 발췌해서 Tbodys 변수에 담았다.

rowList = []
# 정제한 정보를 담을 rowList라는 이름의 비어진 배열을 만들었다.
columnList = []
# 반복문에서 임시로 사용할 columnList라는 이름의 비어진 배열을 만들었다.

rows = Tbodys.findAll(name = "tr")
# Tbodys에서 tag name이 tr인 부분을 발췌해서 rows라는 이름의 배열에 담았다.

columns1 = rows[1].findAll(name = ["th","td"])

# 외부 반복문은 rows의 길이 만큼 반복을 실행
for j in range(0, len(rows)):
#   columns 변수에 rows의 각 인덱스값에서 tagname이 th,td인 부분을 발췌해서 담았다.
    columns = rows[j].findAll(name = ["th","td"])
    # 인덱스값에 내용이 없는 경우는 그냥 넘어간다.
    if columns == []:
        continue
    # 두번째 반복문은 columns의 길이만큼 반복을 실행
    for i in range(0, len(columns)):
#       columnList배열에 columns각 인덱스 값에서 text만을 발췌해서 담아낸다.
        columnList.append( columns[i].text)
    # rowsList(전체리스트)에 columnList 발췌내용 추가
    rowList.append(columnList)
#   columnList를 다시 비워준다.(새로 담아내기 위해서)
    columnList = []

# result에 결과를 DataFrame형태로 변환하여 담아내고 컬럼이름을 변경해주었다
result = pd.DataFrame(rowList, columns = ["날짜","요일","기념일","유형"] )
```
