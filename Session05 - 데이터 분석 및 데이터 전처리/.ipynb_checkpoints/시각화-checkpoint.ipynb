#### 한페이지에 두가지 그래프 그리기

```python
st01Data = pd.read_csv('../dataset/st0001data.csv')
st02Data = pd.read_csv('../dataset/st0002data.csv')

st01Data.head()

x1 = st01Data.WEEK
y1 = st01Data.MEAN_VOLUME

x2 = st02Data.WEEK
y2 = st02Data.MEAN_VOLUME
```

```python
# 도화지만들기
# 10은 x축 5는 y축길이
fig = plt.figure(figsize=(10,5))

# 1행 2열 중 1열에 사용
ax1 = fig.add_subplot(211)
ax1.set_xlabel("WEEK")
ax1.set_ylabel("MEAN_VOLUME")
ax1.plot(x1,y1,"b--", label="st0001")
ax1.legend(loc=0)
ax2 = fig.add_subplot(212)
ax2.set_xlabel("WEEK")
ax2.set_ylabel("MEAN_VOLUME")
ax2.plot(x2,y2,"g", label="st0002")
ax2.legend(loc=0)
```

```python
# 도화지만들기
# 10은 x축 5는 y축길이
fig = plt.figure(figsize=(10,5))

# 1행 2열 중 1열에 사용
ax1 = fig.add_subplot(111)
ax1.set_xlabel("ax1_WEEK")
ax1.set_ylabel("ax1_MEAN_VOLUME")
ax1.plot(x1,y1,"b--", label="st0001")
ax1.legend(loc=0)
ax2 = ax1.twinx()
ax2.set_xlabel("ax2_WEEK")
ax2.set_ylabel("ax2_MEAN_VOLUME")
ax2.plot(x2,y2,"g", label="st0002")
ax2.legend(loc=1)
```

../dataset/feature_regression_example.csv
2015년도 자료만 활용하여
x축: week 이중축으로 qty, pro_percent를
차트로 그리세요

```python
# 판매량의 범위는 500~4000으로 움직이는데 프로모션는 0~0.35로 이루어진다.
```

```python
ex = pd.read_csv('../dataset/feature_regression_example.csv')
```

```python
ex.columns = ex.columns.str.lower()
```

```python
ex2015 = ex.loc[ex.year == 2015]
```

```python
testX1 = ex2015.week
testY1 = ex2015.qty

testX2 = ex2015.week
testY2 = ex2015.pro_percent
```

```python
# 도화지만들기
# 10은 x축 5는 y축길이
fig = plt.figure(figsize=(10,5))

# 1행 2열 중 1열에 사용
ax1 = fig.add_subplot(111)
# ax1.set_xlabel("ax1_WEEK")
ax1.set_ylabel("qty")
ax1.plot(testX1,testY1,"b--")
ax1.legend(loc=0)
ax2 = ax1.twinx()
ax1.plot(testX2,testY2,"g--")
ax2.set_xlabel("WEEK")
ax2.set_ylabel("pro_percent")
ax2.legend(loc=1)
```
