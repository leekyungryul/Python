#### np.where활용

#### 데이터 조작방법
1. np.where 조작한다.(*복잡한 로직에는 힘들어진다.)
2. 함수를 활용하여 조작한다.(*복잡한 로직에서도 강하다)
3. loc를 활용해서 조작한다.(*가독성이 뛰어나나 속도가 떨어진다.)
4. map함수를 사용할 수도 있다.(머신/딥러닝에서 많이 사용된다.)

```python
import numpy as np
customerData = pd.read_csv("../dataset/customerdata.csv")
```

```python
# PRODUCTAGE가 1보다 작은경우 1 아니면 0
np.where(customerData.PRODUCTAGE < 1, 1, 0)
```

```python
# PRODUCTAGE_NEW라는 이름의 컬럼을 생성해서 
# PRODUCTAGE가 1보다 작으면 1 아니면 PRODUCTAGE값을 그대로 표시
customerData["PRODUCTAGE_NEW"] = \
np.where(customerData.PRODUCTAGE < 1, 1, customerData.PRODUCTAGE)
```

```python
customerData
```

```python
# 검증하기
customerData.loc[(customerData.PRODUCTAGE < 1) &
                (customerData.PRODUCTAGE_NEW !=1)]
```

```python
# PRODUCTAGE_NEW라는 이름의 컬럼을 생성해서 
# PRODUCTAGE가 1보다 작으면 1 아니면 PRODUCTAGE가 2보다 작으면 2 아니면 3
customerData["PRODUCTAGE_NEW"] = \
np.where(customerData.PRODUCTAGE < 1, 1, \
         np.where(customerData.PRODUCTAGE < 2, 2, 3))
```

```python
customerData
```

#### 함수활용

함수정의 

1 미만인 경우 = 1

2 미만인 경우 = 2

그외 3

```python
# 함수 생성
def valueClustering(inValue):
    
    if inValue < 1:
        outValue = 1
    elif inValue < 2:
        outValue = 2
    else: 
        outValue = 3
    return outValue
```

```python
valueClustering(1.2)
```

```python
# customerData의 PRODUCTAGE컬럼값에 생성한 함수를 적용한다.
customerData.PRODUCTAGE.apply(valueClustering)
```

```python
# 새로운 컬럼을 만들어서 함수가 적용되어 계산된 컬럼값을 대입한다.
customerData["PRODUCTAGE_NEW"] = customerData.PRODUCTAGE.apply(valueClustering)
```

```python
customerData
```
