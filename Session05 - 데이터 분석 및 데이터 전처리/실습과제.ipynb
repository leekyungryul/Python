productage가 1보다작으면 1 2보다작으면2 3보다작으면3
이외는 5라고 새로운 컬럼 product_age_new를 생성하세요

```python
import pandas as pd
import numpy as np
```

```python
customerData = pd.read_csv("../dataset/customerdata.csv")
customerData
```

```python
customerData["product_age_new"] = np.where(customerData.PRODUCTAGE < 1, 1, \
                                           np.where(customerData.PRODUCTAGE < 2, 2, \
                                                    np.where(customerData.PRODUCTAGE < 3, 3, 5)))
```

```python
customerData
```

```python
customerData.loc[(customerData.PRODUCTAGE < 1) & (customerData.product_age_new != 1)]
```

```python
customerData.loc[(customerData.PRODUCTAGE >= 1) &\
                 (customerData.PRODUCTAGE < 2) &\
                 (customerData.product_age_new != 2)]
```

```python
# 컬럼내에서 중복을 제거한 데이터를 보여준다. 
customerData.CUSTTYPE.drop_duplicates()
```

```python
# Custtype별로 숫자를 맵핑했다.
customerData["CUSTTYPE_NO"] = np.where(customerData.CUSTTYPE == "Big-Screen-lover", "1", \
                                      np.where(customerData.CUSTTYPE == "Sleeping-dog", "2", "3"))
```
