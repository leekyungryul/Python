[데이터 통합]
selloutData 자료에서
YEAR, WEEK 컬럼을 생성하고 WEEK 가 52 이하인
데이터만 조회한 후 refinedSelloutData 변수에 담으세요

```python
import pandas as pd
import numpy as np
```

```python
# selloutData에 pandas처리
selloutData = pd.read_csv("../dataset/kopo_channel_seasonality_new.csv")
selloutData
```

```python
# selloutData의 컬럼들의 값을 형변환처리
selloutData = selloutData.astype({"REGIONID":str,
                                         "PRODUCT":str,
                                         "YEARWEEK":str,
                                         "QTY":float})
```

```python
# YEARWEEK에서 년도정보만 인덱싱하기 위해서 함수생성
def year(inValue):
    outValue = inValue[0:4]
    return outValue
```

```python
# YEARWEEK에서 주차정보만 인덱싱하기 위해서 함수생성
def week(inValue):
    outValue = inValue[4:6]
    return outValue
```

```python
selloutData["YEAR"] = selloutData.YEARWEEK.apply(year)
selloutData["WEEK"] = selloutData.YEARWEEK.apply(week)
# 아래와 같은 방식으로 인덱싱하여도 가능
# selloutData["YEAR"] = selloutData.YEARWEEK.str[0:4]
# selloutData["WEEK"] = selloutData.YEARWEEK.str[4:6]
```

```python
# 52보다 작은 행들을 표시하기위해서 아래와 같이 정수로 형변환해서 계산
refinedSelloutData = selloutData.loc[selloutData.WEEK.astype(int) <= 52]
refinedSelloutData
```

[불량 데이터 처리]
kopo_channel_seasonality_new.csv 자료를 담은
selloutData 변수에서
QTY컬럼 음수(반품)인 경우 0, 양수인 경우 기존 QTY 값
유지하는 로직을 적용하여 QTY_NEW 컬럼을 추가하세요

```python
refinedSelloutData["QTY_NEW"] = \
np.where(refinedSelloutData.QTY.astype(int) < 0, 0, refinedSelloutData.QTY)
refinedSelloutData
```

반복문으로 컬럼들을 문자형으로 변환

```python
# 컬럼들을 리스트로 만들어서 변수에 대입한다.
columnList = list(selloutData.columns)
columnList
```

```python
# 컬럼들이 대입된 리스트의 길이만큼 반복하여 
for i in range(0, len(columnList)):
#     "str"을 변수에 대입하고
    commonType = "str"
#     각 컬럼을 불러내서 위에 변수에 담겨있는 문자형으로 변환하여준다.
    selloutData[columnList[i]] = selloutData[columnList[i]].astype(commonType)
```

refinedSelloutData 에서 →
지역, 상품, 연주차 컬럼순으로 오름차순 정렬하여
sortedData 변수에 담으세요

```python
refinedSelloutData
```

```python
# 아래에 리스트에 담겨있는 컬럼들을 기준으로 오름차순으로 정렬을 한다.
# inplace는 false로 두고 새로운 변수에 대입한다.
sortKey = ["REGIONID","PRODUCT","YEARWEEK"]
sortedData = refinedSelloutData.sort_values(by=sortKey,
#                                      ascending=[True,True,True],
                                     inplace=False)

# 만약 카피를 하지않고 sortedData = refinedSelloutData을 한다면 
# 둘다 같은 주소를 바라보기 때문에 서로연결되어서 각각 분리된 사용이 불가하다.
# sortedData = refinedSelloutData.copy()
```

[지역, 상품, 연도 별 집계]
sortedData 에서 지역, 상품, 연도 단위
판매량(QTY_NEW) 의 평균 연산 후
groupData 변수에 담으세요
이후 컬럼명을 QTY_MEAN로 변경하세요

```python
# 지역,상품,연도단위별로 그룹을 묶어서 판매량 평균을 새로운 컬럼에 대입한다.
# 연도주차별로 구분되어있던 데이터가 년도별로 묶였으므로 행이 1/52만큼 줄었다
groupKey = ["REGIONID","PRODUCT","YEAR"]
groupData = sortedData.groupby(groupKey)[["QTY_NEW"]].agg("mean")
groupData = groupData.rename(columns = {'QTY_NEW':'QTY_MEAN'})
```

#### 이동집계함수

```python
sortedData.head(10)
```

```python
sortedData["MA5"] = sortedData.QTY.rolling\
                    (window=5, 
                     min_periods=1, 
                     center=True).mean()
```

```python
sortedData.head(10)
```

```python
# min_periods를 제거하면 계산에 필요한 행의 갯수가 설정치인 5개에 미달하면 nan으로 반환한다.
sortedData["MA5_BASIC"] = sortedData.QTY.rolling\
                    (window=5, 
#                      min_periods=1, 
                     center=True).mean()
```

```python
sortedData.head(10)
```

```python
# center를 false로 두게되면 현재행을 기준으로 앞에서부터 현재 데이터까지의 평균이 계산되게 된다.
# 아래는 min_periods를 1로 두게 되었으므로 첫번째 값은 본인값이 그대로 반환된다.
sortedData["MA5_BASIC_NO"] = sortedData.QTY.rolling\
                            (window=5,
                            min_periods=1,
                            center=False).mean()
```

```python
sortedData.head(10)
```

```python
sortedData["EFFECT"] = sortedData.QTY_NEW / sortedData.MA5
```

```python
# 년도의 중첩을 제거하고 주차별로만 그룹으로 묶었다.
groupKey = ["REGIONID","PRODUCT","WEEK"]
```
